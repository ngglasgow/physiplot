```python
xpixels = 200
ypixels = 100
aspect_ratio = ypixels / xpixels
frames = 110
images = [np.random.rand(ypixels, xpixels) for i in range(frames)]
image_arr = np.array(images)

x = range(image_arr.shape[2])
y = range(image_arr.shape[1])
mean_image = image_arr.mean(axis=0)
```

```python
image_arr = io.imread('image.tif')
x = range(image_arr.shape[2])
y = range(image_arr.shape[1])
mean_image = image_arr.mean(axis=0)
frames = image_arr.shape[0]
```

```python
play = widgets.Play(interval=10, value=0, min=0, max=frames-1, step=1, description='Frame')
slider = widgets.IntSlider(min=0, max=frames-1, description='Frame')
widgets.jslink((play, 'value'), (slider, 'value'))
# roi_button = widgets.Button(description='Draw ROIs')

plot = figure(plot_width=600, plot_height=int(600*aspect_ratio),
              match_aspect=True,
              x_range=(0, xpixels), y_range=(0, ypixels),
              toolbar_location='above')
plot.axis.visible = False
plot.grid.visible = False
image = plot.image(image=[image_arr[0]], x=0, y=0, dw=xpixels, dh=ypixels)

def frame_changed(change):
    new_frame = slider.value
    image.data_source.data['image'] = [image_arr[new_frame]]
    push_notebook(handle=h)
    
slider.observe(frame_changed, 'value')
# roi_button.on_click(draw_rois(mean_image))
h = show(plot, notebook_handle=True)

widgets.HBox([play, slider])
```

```python

img = hv.Image(data=(x, y, mean_image))
img.opts(aspect='equal', toolbar='above', cmap='Greys', yaxis=None, xaxis=None)

background = hv.Polygons([])
signal = hv.Polygons([])

background_stream = streams.Stream.define('bg_data', bg_data=streams.PolyDraw(source=background, drag=True, styles={
                                     'fill_color': ['red']},
                                     show_vertices=True,
                                     num_objects=1, linked=True,
                                     name='background'))
signal_stream = streams.Stream.define('sig_data', sig_data=streams.PolyDraw(source=signal, drag=True, styles={
                                 'fill_color': ['green']},
                                 show_vertices=True,
                                 num_objects=1, linked=True,
                                 name='signal'))

background.opts(opts.Polygons(fill_alpha=0.5, active_tools=['poly_draw']))
signal.opts(opts.Polygons(fill_alpha=0.5, active_tools=['poly_draw']))



img * background * signal
```

```python
from holoviews import streams
pointer = streams.PointerXY(x=0, y=0) # x=0 and y=0 are the initialized values

def crosshair(x, y):
    return  hv.Ellipse(0,0,1) * hv.HLine(y) * hv.VLine(x)

hv.DynamicMap(crosshair, streams=[pointer])
```
